1. Target Variable & Quality Insights
Balance of Data: The dataset is well-balanced regarding Tool Condition (roughly 50% worn, 50% unworn).


Yield Rate: While most parts pass visual inspection (Value 1), a significant portion of failures (Value 0) occurs specifically when the tool_condition is 0 (Worn).

2. Sensor "Red Flags" (Predictive Indicators)
By comparing the signals of Worn vs. Unworn tools, we identified the most important features for predictive maintenance:

Current Feedback (X1, Y1, S1): This is our First indicator.

Unworn Tools: Show a steady, predictable current flow.

Worn Tools: Show "noisy" signals with frequent spikes. A worn tool creates more friction, forcing the motor to draw irregular amounts of current to maintain speed.

Output Power (X1_OutputPower): Worn tools consistently demand higher peak power. If X1_OutputPower exceeds a certain threshold, it is a 90% reliable sign that the tool is nearing the end of its life.

Actual vs. Command Gap: In unworn tools, the ActualPosition and CommandPosition are almost identical. In worn tools, we start to see a "lag" or "deviation" during high-speed cutting, indicating that the tool's physical degradation is affecting the machine's precision.

3. Machining Process Analysis
Critical Stages: The "Layering" and "Cutting" processes are the most sensitive. Almost all visual inspection failures happen during these stages.

Safe Stages: Processes like "Prep" and "End" show very little difference in sensor data between worn and unworn tools.



4. Correlation & Multicollinearity
Redundancy: The "Command" columns (Velocity, Acceleration, Position) are mathematically derived and perfectly correlated with each other. For a machine learning model, you can likely drop the "Command" columns and focus only on the "Actual" sensor readings and "Current Feedback" to reduce noise.

Spindle Speed (S1): Spindle velocity is generally kept constant, but the S1_CurrentFeedback increases as the tool wears, even if the velocity stays the same.
